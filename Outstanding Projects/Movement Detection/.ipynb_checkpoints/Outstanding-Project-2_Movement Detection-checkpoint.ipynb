## MOVEMENT DETECTION
We have a video of vehivles moving on a highway, and we will be tring to detect and outline the vehicles moving in the video. We will be using opencv to read videos and images and OS library to access them. The  basic idea is to divide videos into its frame and find the difference in pixels beween consecutive frames ( assuming that the camera is still and the background remains constant)

###### Importing Libraries

```python
import os
import re
import cv2 
import numpy as np
from os.path import isfile, join
import matplotlib.pyplot as plt
```

#### 1) Reading the video
We will create a VideoCapture object and read from input file. If you want to take input from camera, pass 0 instead of the video file name

```python
cap_vid = cv2.VideoCapture('D:/Outstanding Project/Car Movement/input.mp4')

# Check if camera opened successfully
if (cap_vid.isOpened()== False): 
    print("Error opening video stream or file")

# Read until video is completed
while(cap_vid.isOpened()):
    # Capture frame-by-frame
    ret, frame = cap_vid.read()
    if ret == True:
        # Display the resulting frame
        cv2.imshow('Frame',frame)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    # Break the loop
    else: 
        break
        
# When everything done, release the video capture object
cap_vid.release()
# Close all the frames
cv2.destroyAllWindows()
```

### 2) Preprocessing 
#### Converting video into frames/distinct images
We will try to read the frames as images. Success variable will contain a boolean value informing us if the image was read succesfully or not. While success is true we will continue to save iages in the form of their frame numbers in a folder called as frame

```python
cap_vid = cv2.VideoCapture('D:/Outstanding Project/Car Movement/input.mp4')
count = 0
success,image = cap_vid.read()
while success:
    cv2.imwrite("D:/Outstanding Project/Car Movement/frames/{}.png".format(count), image)     # save frame 
    success,image = cap_vid.read()
    if cv2.waitKey(10) == 27:                     # exit if Escape is hit
        break
    print("Reading:{}".format(count)+"\n")
    count += 1
```

#### Accessing Frames

```python
# get file names of the frames
col_frames = os.listdir('D:/Outstanding Project/Car Movement/frames')

# sort file names to make sure they are in the correct order
col_frames.sort(key=lambda f: int(re.sub('\D', '', f)))

# empty list to store the frames
col_images=[]

for i in col_frames:
    # read the frames
    img = cv2.imread('frames/'+i)
    # append the frames to the list
    col_images.append(img)
```

### 3) Capturing movement in video
We will take two consecutive frames, convert them into greyscale and subtract the pixel values at each pixel. This concept is known as Frame differencing. It may seem like there isnt any diference between both frames after following the process we will be able to detect a slight change or movement. 
###### To exagerate the difference we will be using image thresholding
In this technique, the pixel values of a grayscale image are assigned one of the two values representing black and white colors based on a threshold.So, if the value of a pixel is greater than a threshold value, it is assigned one value, else it is assigned the other value.
###### Threshold in this case is (30,255)

##### Data Exploration

```python
# plot 13th frame
i = 13

for frame in [i, i+1]:
    plt.imshow(cv2.cvtColor(col_images[frame], cv2.COLOR_BGR2RGB))
    plt.title("frame: "+str(frame))
    plt.show()
```

##### Image Differencing

```python
# convert the frames to grayscale
grayA = cv2.cvtColor(col_images[i], cv2.COLOR_BGR2GRAY)
grayB = cv2.cvtColor(col_images[i+1], cv2.COLOR_BGR2GRAY)

# plot the image after frame differencing
plt.imshow(cv2.absdiff(grayB, grayA), cmap = 'gray')
plt.show()
```

##### Image Thresholding

```python
diff_image = cv2.absdiff(grayB, grayA)

# perform image thresholding
ret, thresh = cv2.threshold(diff_image, 30, 255, cv2.THRESH_BINARY)

# plot image after thresholding
plt.imshow(thresh, cmap = 'gray')
plt.show()
```

### 4) Adding Blur (Gaussian and Dialation)
There is still scope of improvement. The idea is to merge the nearby white regions to have fewer contours and for that, we can use techniques like Gaussian Blur and Image Dialation. 
While Gaussian Blur smmothens the image, Image Dialation blends some unnecessary fragments of the white region to give better contours

```python
blur = cv2.GaussianBlur(thresh,(3,3),0)
```

```python
plt.imshow(blur, cmap = 'gray')
plt.show()
```

```python
# apply image dilation
kernel = np.ones((3,3),np.uint8)
dilated = cv2.dilate(blur,kernel,iterations = 1)

# plot dilated image
plt.imshow(dilated, cmap = 'gray')
plt.show()
```

For the purpose of simplification, we are not going to use the entire frame to detect moving vehicles. We will first select a zone, and if a vehicle moves into that zone, then only it will be detected.
The zone in this case is the part of the image below the line.

```python
plt.imshow(dilated)
cv2.line(dilated, (0, 80),(256,80),(100, 0, 0),5)
plt.show()
```

### Contours
I have used the word Contours before. Contours are used to identify the shape of an area in the image having the same color or intensity. Contours are like boundaries around regions of interest. To mark the moving vehiclees we need their contours and we will be extracting them from out dialated threhold image

```python
# find contours
contours, hierarchy = cv2.findContours(thresh.copy(),cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
```

```python
valid_cntrs = []

for i,cntr in enumerate(contours):
    x,y,w,h = cv2.boundingRect(cntr)
    if (x <= 200) & (y >= 80) & (cv2.contourArea(cntr) >= 25):
        valid_cntrs.append(cntr)

# count of discovered contours        
len(valid_cntrs)
```

### Looking at Our Contours

```python
dmy = col_images[13].copy()

cv2.drawContours(dmy, valid_cntrs, -1, (127,200,0), 2)
cv2.line(dmy, (0, 80),(256,80),(100, 255, 255))
plt.imshow(dmy)
plt.show()
```

### 5)Find the movement in video
After analyzing the data and viewing the results of out process, we can move on to detecting the movement for the whole video. The entire process remains the same. The only difference is that the process is repeated for all continious frames
### 6) Extracting Contours
After detecting the video we have extracted contours and added the outline to each individual frame. The frames are saved in a folder named output.

```python
# kernel for image dilation
kernel = np.ones((4,4),np.uint8)

# font style
font = cv2.FONT_HERSHEY_SIMPLEX

# directory to save the ouput frames
pathIn = "D:/Outstanding Project/Car Movement/output/"

for i in range(len(col_images)-1):
    
    # frame differencing
    grayA = cv2.cvtColor(col_images[i], cv2.COLOR_BGR2GRAY)
    grayB = cv2.cvtColor(col_images[i+1], cv2.COLOR_BGR2GRAY)
    diff_image = cv2.absdiff(grayB, grayA)
    
    # image thresholding
    ret, thresh = cv2.threshold(diff_image, 30, 255, cv2.THRESH_BINARY)
    
    #GAussian Blur
    blur = cv2.GaussianBlur(thresh,(3,3),0)
    
    # image dilation
    dilated = cv2.dilate(blur,kernel,iterations = 1)
    
    # find contours
    contours, hierarchy = cv2.findContours(dilated.copy(), cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
    
    # shortlist contours appearing in the detection zone
    valid_cntrs = []
    for cntr in contours:
        x,y,w,h = cv2.boundingRect(cntr)
        if (x <= 200) & (y >= 80) & (cv2.contourArea(cntr) >= 25):
            if (y >= 90) & (cv2.contourArea(cntr) < 40):
                break
            valid_cntrs.append(cntr)
            
    # add contours to original frames
    dmy = col_images[i].copy()
    cv2.drawContours(dmy, valid_cntrs, -1, (127,200,0), 2)
    
    cv2.putText(dmy, "vehicles detected: " + str(len(valid_cntrs)), (55, 15), font, 0.6, (0, 180, 0), 2)
    cv2.line(dmy, (0, 80),(256,80),(100, 255, 255))
    cv2.imwrite(pathIn+str(i)+'.png',dmy)  
```

### Creating a Video out of the output Frames

```python
# specify video name
pathOut = 'vehicle_detection_v3.mp4'

# specify frames per second
fps = 14.0
```

```python
frame_array = []
files = [f for f in os.listdir(pathIn) if isfile(join(pathIn, f))]
```

```python
files.sort(key=lambda f: int(re.sub('\D', '', f)))
##  sorting frames to ensure that they are in the correct order
for i in range(len(files)):
    filename=pathIn + files[i]
    print(filename)
    #read frames
    img = cv2.imread(filename)
    height, width, layers = img.shape
    size = (width,height)
    print(size)
    #inserting the frames into an image array
    frame_array.append(img)
```

### Writing the video Vehicle_detection

```python
out = cv2.VideoWriter(pathOut,cv2.VideoWriter_fourcc(*'DIVX'), fps,(256,144))

for i in range(len(frame_array)):
    # writing to a image array
    out.write(frame_array[i])

out.release()
```
